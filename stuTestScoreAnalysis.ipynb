### Project 2: Student Test Score Analysis
#### Objective: Visualize test score distributions for students from different schools and identify top-performing groups.

##### Step 1: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

##### Step 2: Create a sample dataset

```python
data = {
    'school': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'score': [78, 85, 90, 70, 75, 68, 88, 92, 95, 80]
}

students = pd.DataFrame(data)
```

##### Step 3: View the data

```python
print(students)
```

##### Step 4: Histogram - Overall score distribution

```python
# Histogram
sns.histplot(students['score'], bins=5, kde=True)

# KDE curve separately
# sns.kdeplot(students['score'], color='red', linewidth=2)

# sns.histplot(
#     students['score'], 
#     bins=5, 
#     kde=True,
#     fill=True, 
#     color='lightblue',          # color of histogram bars
#     line_kws={'color':'red', 'linewidth':2}  # color and thickness of KDE curve
# )

plt.title('Overall Test Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()

# sns.histplot() → makes a histogram to show how scores are distributed
# bins=5 → divides the scores into 5 ranges (bins).
# kde=True → adds a smooth curve showing the overall shape of the data.Kernel Density Estimate (KDE)
```

##### Step 5: Box Plot - Compare scores by school

```python
sns.boxplot(x='school', y='score', data=students)
plt.title('Test Score Comparison by School')
plt.xlabel('School')
plt.ylabel('Score')
plt.show()

# Each box shows how scores are spread (min, max, median, etc.)-black line
# Helps identify which school performs best and if any have outliers
```
