##### Step 1: Import libraries

```python
import matplotlib.pyplot as plt
```

##### Step 2: Sample data

```python
x = [1, 2, 3, 4, 5, 6]
y = [10, 15, 13, 20, 18, 25]
```

##### Step 3: Create subplots (1 row, 2 columns), Line chart on first subplot, Scatter plot on second subplot

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # axes[0] and axes[1], width=12, height = 5

# -----------------------Line chart on first subplot-----------------------------------

axes[0].plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2)
axes[0].set_title('Line Chart', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].grid(True)

# -------------------------------Scatter plot on second subplot--------------------------------

axes[1].scatter(x, y, color='red', marker='s', s=100)  # s = size of marker
axes[1].set_title('Scatter Plot', fontsize=14, fontweight='bold')
axes[1].set_xlabel('X-axis')
axes[1].set_ylabel('Y-axis')
axes[1].grid(True)

#  -------------------Adjust layout and show figure---------------------------------------

plt.tight_layout()
plt.show()
```
