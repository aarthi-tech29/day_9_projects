### Project 2: Titanic Survival Analysis
#### Objective: Use a real-world dataset to visualize survival rates based on categorical features. 

##### Step 1: Import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Step 2: Load the Titanic dataset

```python
titanic = sns.load_dataset('titanic')

# directly loads the Titanic dataset built into Seaborn.
```

##### Step 3: Show first few rows

```python
print(titanic.head())

# .head() shows the first 5 rows of the dataset.
```

##### Step 4: Survival Bar Plot

```python
# Shows average survival rate by passenger class
sns.barplot(x='class', y='survived', data=titanic)
# sns.barplot(x='class', y='survived', data=titanic, errorbar=None)

plt.title('Average Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# x='class' - each bar represents a passenger class (First, Second, Third).
# y='survived' - shows the average survival rate (1 means survived, 0 means not).
# data=titanic - automatically calculates the mean survival rate for each class and plots it.
# errorbar=None removes the black line (no confidence interval shown)
# the bar for First Class shows survival rate = 0.63,and the black line goes from 0.55 to 0.70.
# 95% confident the true average survival rate for First Class passengers is between 55% and 70%
```

##### Step 5: Survival by Age and Gender (Violin Plot)

```python
# Survival by Age and Gender
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, split=True)
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()

# x='sex' - divides passengers by male and female.
# y='age' - shows the age of passengers.
# hue='survived' - separates survivors (1) and non-survivors (0) with different colors.
# split=True - draws the survivor and non-survivor shapes side by side in one violin for easy comparison.
```

##### Step 5: Survival by Embarkation Town


```python
sns.barplot(x='embarked', y='survived', data=titanic, errorbar=None)
plt.title('Survival Rate by Embarkation Town')
plt.xlabel('Embarkation Town (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Survival Rate')
plt.show()

# x='embarked' - uses the town where passengers boarded the ship (C, Q, or S).
# y='survived' - shows the average survival rate for passengers from each town.
```
